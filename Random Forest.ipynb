```python
import pandas as pd
import numpy as np
import openpyxl
import seaborn as sb
import seaborn as sns
import matplotlib.pyplot as plt
import pickle
import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder, OneHotEncoder, OrdinalEncoder

#for classfication
from sklearn.ensemble import RandomForestClassifier

# for resampling 
from imblearn.over_sampling import RandomOverSampler

# for checking testing results
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score, classification_report, confusion_matrix
```

```python
df = pd.read_csv('Energy data 1990 - 2020.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

# Preprocessing

```python
df.isnull().sum()
```

```python
#menghilangkan variabel yang tidak dibutuhkan
df=df.drop(columns=['Share of wind and solar in electricity production (%)'], inplace=False,axis=1)
df=df.drop(columns=['Coal and lignite production (Mt)'], inplace=False,axis=1)
df=df.drop(columns=['Region'], inplace=False,axis=1)
df=df.drop(columns=['Natural gas production (bcm)'], inplace=False,axis=1)
df=df.drop(columns=['Coal and lignite domestic consumption (Mt)'], inplace=False,axis=1)
df=df.drop(columns=['Year'], inplace=False,axis=1)
```

```python
df
```

```python
#menghitung jumlah output
df['country'].value_counts().sort_values(ascending = False)
```

```python
#melabelkan data output
df[['country']] = df[['country']].apply(LabelEncoder().fit_transform)

#view udpated DataFrame
print(df)
```

```python
#menghitung jumlah output
df['country'].value_counts().sort_values(ascending = False)
```

```python
#feature selection data and split data
X = df.drop(columns='country')
y = df['country']
```

```python
#data traing 90% dan data testing 10% before random oversampling
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=200)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#Transformasi/normalisasi data dengan MinMax before resampling
columns = ['CO2 emissions from fuel combustion (MtCO2)', 'Average CO2 emission factor (tCO2/toe)', 'CO2 intensity at constant purchasing power parities (kCO2/$15p)', 'Total energy production (Mtoe)', 'Total energy consumption (Mtoe)', 'Share of renewables in electricity production (%)', 'Share of electricity in total final energy consumption (%)']
scaler = MinMaxScaler()
X_train_scale = scaler.fit_transform(X_train[columns])
X_test_scale = scaler.transform(X_test[columns])
```

```python
X_train[columns]=X_train_scale
X_test[columns]=X_test_scale
```

# RFC

```python
seed=123
np.random.seed(seed)
```

```python
rfc_100 = RandomForestClassifier()

# fit the model to the training set

rfc_100.fit(X_train, y_train)

# Predict on the test set results

y_pred_100 = rfc_100.predict(X_test)
```

```python
# Check accuracy score 
print('Model accuracy score with 100 random forest : {0:0.4f}'. format(accuracy_score(y_test, y_pred_100)))
```

```python
cm = confusion_matrix(y_test, y_pred_100)

print('Confusion matrix\n\n', cm)
```

```python
print ('accuracy_score: ',accuracy_score(y_test, y_pred_100))
print ('f1_score: ',f1_score(y_test, y_pred_100, pos_label='positive', average='micro'))
print ('recall_score: ',recall_score(y_test, y_pred_100, pos_label='positive', average='micro'))
print ('precision_score: ',precision_score(y_test, y_pred_100, pos_label='positive', average='micro'))
```
